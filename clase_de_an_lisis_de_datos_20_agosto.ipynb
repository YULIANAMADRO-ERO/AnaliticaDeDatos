```python
df_o_genero = pd.merge(df_o, df_cgsse[['DIRECTORIO', 'SECUENCIA_P', 'ORDEN','P3271']],
                        on=['DIRECTORIO', 'SECUENCIA_P', 'ORDEN'], how='left')
```

```python
#Calculo de hombres ocupados
dh = df_o_genero[df_o_genero['P3271'] == 1]['FEX_C18'].sum()

#Calculo de mujeres ocupadas

dm = df_o_genero[df_o_genero['P3271'] == 2]['FEX_C18'].sum()

print (f'Hombres en empleo: {int(dh)}')
print (f'Mujeres en empleo: {int(dm)}')
```

```python
df_d_genero = pd.merge(df_d, df_cgsse[['DIRECTORIO', 'SECUENCIA_P', 'ORDEN','P3271']],
                        on=['DIRECTORIO', 'SECUENCIA_P', 'ORDEN'], how='left')
```

```python
#Calculo de hombres desocupados
dh = df_d_genero[df_d_genero['P3271'] == 1]['FEX_C18'].sum()

#Calculo de mujeres desocupadas

dm = df_d_genero[df_d_genero['P3271'] == 2]['FEX_C18'].sum()

print (f'Hombres en desempleo: {int(dh)}')
print (f'Mujeres en desempleo: {int(dm)}')
```

```python
df_f_genero = pd.merge(df_ft, df_cgsse[['DIRECTORIO', 'SECUENCIA_P', 'ORDEN','P3271']],
                        on=['DIRECTORIO', 'SECUENCIA_P', 'ORDEN'], how='left')
```

```python
#Calculo de hombres en fuerza de trabajo
dh = df_f_genero[df_f_genero['P3271'] == 1]['FEX_C18'].sum()

#Calculo de mujeres fuerza de trbajo

dm = df_f_genero[df_f_genero['P3271'] == 2]['FEX_C18'].sum()

print (f'Hombres en fuerza de trabajo: {int(dh)}')
print (f'Mujeres en fuerza de trabajo: {int(dm)}')
```

```python
# Filtrar solo el Quindío (DPTO = 63)
df_quindio = df_f_genero[df_f_genero['DPTO'] == 63]
```

```python
# Hombres desempleados en Quindío
h_desemp_quindio = df_quindio[(df_quindio['P3271'] == 1) & (df_quindio['FEX_C18'] == 1)]['FEX_C18'].sum()
```

```python
# Mujeres desempleadas en Quindío
m_desemp_quindio = df_quindio[(df_quindio['P3271'] == 2) & (df_quindio['FEX_C18'] == 1)]['FEX_C18'].sum()

print(f"Hombres desempleados en Quindío: {int(h_desemp_quindio)}")
print(f"Mujeres desempleadas en Quindío: {int(m_desemp_quindio)}")
```

```python
# Filter df_d_genero for Quindío (DPTO == 63)
df_desempleo_quindio = df_d_genero[df_d_genero['DPTO'] == 63]

# Calculate unemployed men in Quindío
h_desemp_quindio = df_desempleo_quindio[df_desempleo_quindio['P3271'] == 1]['FEX_C18'].sum()

# Calculate unemployed women in Quindío
m_desemp_quindio = df_desempleo_quindio[df_desempleo_quindio['P3271'] == 2]['FEX_C18'].sum()

print(f"Hombres desempleados en Quindío: {int(h_desemp_quindio)}")
print(f"Mujeres desempleadas en Quindío: {int(m_desemp_quindio)}")
```
